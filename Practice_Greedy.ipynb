모험가 길드

공포도가 X인 모험가는 반드시 X명 이상으로 구성해야 여행을 떠날 수 있다.

입력조건<br>
첫째 줄에 모험가의 수 N이 주어집니다.<br>
둘째 줄에 각 모험가의 공포도의 값을 N이하의 자연수로 주어지며, 각 자연수는 공백으로 구분합니다.

출력조건<br>
여행을 떠날 수 있는 그룹 수의 최댓값을 출력합니다.

입력예시<br>
5<br>
2 3 1 2 2

출력예시<br>
2


```python
N = int(input())
gongpo_list=list(map(int,input().split()))

gongpo_list.sort()

count=0
temp = []

for i in range(N):
  temp.append(gongpo_list[i])
  if(max(temp)==len(temp)):
    temp = []
    count +=1

print(count)
```

곱하기 혹은 더하기

0부터 9까지 이루어진 문자열 S가 주어졋을때 곱하기 더하기 연산을 하여 가장 큰수를 만들어내라

입력조건<br>
첫째 줄에 여러 개의 숫자로 구성된 하나의 문자열 S가 주어진다<br>

출력조건<br>
첫째 줄에 만들어질 수 잇는 가장 큰 수를 출력합니다.

입력예시<br>
02984<br>

출력예시 <br>
576

```python
S = str(input())
result = 0
numList = []
for i in range(len(S)):
  numList.append(int(S[i:i+1]))
  
numList.sort()

for j in numList:
  if result ==0:
    result +=j
  else:
    result *=j

print(S)
print(numList)
print(result)
```

문자열 뒤집기

0과 1로만 이루어진 문자열 S를 받는다.<br>
문자열 S에 잇는 모든 숫자를 전부 같게 만들려고 한다. 할수있는 행동은 연속된 하나 이상의 숫자를 잡고 모두 뒤집는게 가능<br>
예시 S=0001100일때, 연속된 4,5번 을 한번에 뒤집으면 모든 숫자가 0이 되므로 최소 출력값은 1이다

입력조건<br>
첫째 줄에 0과 1로만 이루어진 문자열 S가 주어집니다.

출력조건<br>
첫째 줄에 다솜이가 해야 하는 행동의 최소 횟수를 출력합니다.


```python
data = input()
temp=int(data[0])
count0=0
count1=0

if temp==0:
  count0 =1
else:
  count1 =1

for i in range(1,len(data)):
  if temp == int(data[i]):
    continue
  else:
    temp = int(data[i])
    if int(data[i])==0:
      count0 +=1
    else:
      count1 +=1

print(min(count0,count1))
```

만들 수 없는 금액

동네 편의점의 주인인 동빈이는 N개의 도저을 가직욌다.이때 동전을 이용하여 만들 수 없는 양의 정수 금액 중 최솟값을 구하는 프로그램을 작성하시오.

입력조건<br>
첫째 줄에는 동전의 개수를 나타내는 양의정수 N이 주어집니다.<br>
둘째 줄에는 각 동전의 화폐 단위를 나타내는 N개의 자연수가 주어지며, 각 자연수는 공백으로 구분합니다.

출력조건<br>
첫재 줄에 주어진 동전들로 만들 수 없는 양의 정수 금액 중 최솟값을 출력합니다.

입력예시<br>
5<br>
3 2 1 1 9



```python
N=int(input())
coinList = list(map(int,input().split()))
coinList.sort(reverse=True)
target = 0
temp =0

while True:
  if(temp != 0):
    print("target",target)
    break
  target +=1
  temp=target

  for i in coinList:
    if temp-i ==0:
      temp -=i
      break
    elif temp-i <0:
      continue
    else:
      temp -=i
```
