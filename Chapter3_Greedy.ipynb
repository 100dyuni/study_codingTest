당신은 음식점의 계산을 도와주는 점원이다. 카운터에는 거스름을 사용할 500원,100원, 50원, 10원 짜리 동전이 무한히 존재한다고 가정한다, 손님에게 거슬러 줘야 할 돈이 n원일 때 거슬러 줘야 할 동전의 최소 개수를 구하라


```python
n = 1260
count = 0
#큰 단위부터 확인
coin_type = [500,100,50,10]

for coin in coin_type:
  count += n // coin
  n %= coin

print(count)
```

큰 수의 법칙
배열이 있을때 주어진 수들을 M번 더하여 가장 큰 수를 만드는 법칙. 단 해당하는 수가 연속해서 k번을 초과하여 더해질수 없게 만들것

배열의 크기 N, 숫자가 더해지는 횟수 M, 그리고 K가 주어질 때 큰수의 법칙에 따른 결과를 출력하시오.

##입력조건
1. 입력받는 수는 자연수로 공백으로 구분한다. N M K
2. 배열의 숫자도 자연수로 공백으로 입력받는다.
3. k는 항상 m보다 작거나 같다.

```python
#공백을 이용하여 각각의 변수 받기
N, M, K = map(int, input().split())
data = list(map(int, input().split()))

data.sort() #입력받은 변수 정렬

first = data[N-1]
second = data[N-2]

result =0
# N배열길이, M더할횟수, K동일숫자 더할회수
while True:
  for i in range(K):
    if M==0:
      break;
    result += first
    M -= 1
  
  if M ==0:
    break;
  result += second
  M -=1

print(result)
```

```python
#공백을 이용하여 각각의 변수 받기
N, M, K = map(int, input().split())
data = list(map(int, input().split()))

data.sort() #입력받은 변수 정렬

first = data[N-1]
second = data[N-2]

result =0

#반복문을 계속해서 돌릴경우 시간적인 요소가 너무 많이 사용하게 됨.
#수열을 이용하여 해당 문제를 해결하는 방안을 구축
#가장 큰 수가 더해지는 횟수 계산
count = int(M/(K+1)) * K
count += M%(K+1)

result += first * count
result += second * (M-count)

print(result)
```

##숫자 카드게임
1. 숫자가 쓰인 카드들이 N * M 형태로 놓아져있다. N 행, M 열
2. 뽑고자 하는 행을 선택
3. 선택된 행에 포함된 카드들 중 가장 낮은 카드를 뽑아야한다,
4. 이후 다른행의 가장 낮은 카드와 비교하였을때 그중에서 제일 큰 값을 출력


##입력조건
1. 행과 열은 공백 기준으로  받는다
2. 둘째 줄 부터 N개의 줄에 걸쳐 각 카드에 적힌 숫자가 주어진다.


```python
# N,M을 공백 구분하여 받기
N, M = map(int, input().split())

result =0

for i in range(N):
  datalist = list(map(int,input().split()))
  minVal = min(datalist)
  result = max(minVal,result)

print(result)
```

## 1이 될 때까지
어떠한 수 N이 1이 될 때까지 다음의 두 과정 중 하나를 반복적으로 수행하여 수행하려고 한다. 두번째 연산은 N이 K로 나누어떨어질 때만 선택할 수 있다.
1. N에서 1을 뺀다
2. N을 K로 나눈다.

입력예시
25 5

출력예시
2

```python
# N,M값 받아오기
N, M = map(int,input().split())
count = 0
while True :
  if N<M:
    break
  
  target = (N//M)*M
  count += (N - target)
  N = target

  count += 1
  N //=M

  print(str(N) + " : " +str(count))
  
print(count)
```
