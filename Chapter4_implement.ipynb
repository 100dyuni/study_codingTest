4-1) 상하좌우
입력조건
1. 첫째 줄에 공간의 크기를 나타내는 N이 주어진다 N*N공간 구현
2. 둘째 줄에 여행가 A가 이동할 계획서 내용이 주어진다
3. 최종적으로 도착할 곳의 좌표를 출력한다.


```python
N = int(input())
x,y =1,1
plans = input().split()

dx = [0,0,-1,1]
dy = [-1,1,0,0]
move_types= ['L','R','U','D']

for plan in plans:
  for i in range(len(move_types)):
    if plan == move_types[i]:
      nx = x+dx[i]
      ny = y+dy[i]

      if nx<1 or ny<1 or nx>N or ny>N:
        continue

      x,y = nx,ny

print(x,y)
```

1. 첫째 줄에 정수 N이 입력된다.

4-3) 왕실 나이트
8*8 체스판에서 조표값을 주어주고 나이트 움직일수 있는 경우의 수를 구하시오

입력조건
1. 현재 나이트가 위치한 곳의 좌표를 나타내는 두 문자를 구성된 문자열이 입력된다.


```python
input_data = input()
row = int(input_data[1])
#ord(c)는 문자의 유니코드 값을 돌려주는 함수이다.
column = int(ord(input_data[0])) - int(ord('a')) +1

#나이트가 이동할수 있는 방향 정의
steps = [(-2,-1),(-2,+1),(-1,-2),(-1,+2),(1,2),(1,-2),(2,1),(2,-1)]

count =0

for step in steps:
  next_row = row + step[0]
  next_col = column + step[1]

  if next_row>=1 and next_row <=8 and next_col >=1 and next_col <=8:
    count += 1

print(count)
```

4-4) 게임개발
시뮬레이션 문제

입력조건.<br>
첫째 줄에 맵의 세로 크기 N과 가로크기 M을 공백으로 구분하여 입력한다.<br>
둘째 줄에 게임 케릭터가 있는 칸의 좌표 (A,B)와 바라보는 방향 d가 각각 서로 공백으로 구분하여 주어진다. 방향 d의 값으로는 다음과 같이 4가지가 존재한다..<br> 
 0 북/1 동/2 남/3 서 .<br>
셋째 줄 부터 맵이 윤지인지 바다인지 정보가 주어진다..<br>
 0 육지/1 바다.<br>
처음 게임 캐릭터가 위치한 칸의 상태는 항상 육지이다.

```python
N,M = map(int,(input().split()))

#맵 생성 방문한 위치를 저장히기 위한 맵을 생성하여 0으로 초기화
d = [[0] * M for _ in range(N)]
x,y,direction = map(int,(input().split()))
d[x][y] =1

array = []
for i in range(N):
  array.append(list(map(int,input().split())))

#북, 동, 남, 서 방향 정의
dx = [-1, 0 ,1 ,  0]
dy = [0 , 1 ,0 , -1]

#방향회전 북 0, 동 1, 남 2, 서 3
def turn_left():
  global direction
  direction -=1
  if direction == -1:
    direction =3

#시뮬레이션 시작
count =1
turn_time = 0
while True:
  turn_left()
  nx = x + dx[direction]
  ny = y + dy[direction]

  if d[nx][ny] == 0 and array[nx][ny] ==0:
    d[nx][ny] =1
    x = nx
    y = ny
    count += 1
    turn_time =0
    continue
  else:
    turn_time +=1

  #네방향 모두 갈수 없는 경우
  if turn_time ==4:
    nx = x - dx[direction]
    ny = y - dy[direction]
    #뒤로 갈수 있다면 이동하기
    if d[nx][ny] == 0 and array[nx][ny] == 0:
      x= nx
      y= ny
      trun_time = 0
    else:
      break

print(count)
```
